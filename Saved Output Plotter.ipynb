```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pickle

%matplotlib inline
```

```python
with open('CA1_pyr_cell_LC.pkl', 'rb') as f: 
    data = pickle.load(f)
    CA1_TS  = data[0]
    CA1_Spikes = data[1]
    CA1_SpikeIndex = data[2]
    CA1_Som_Current = data[3]
    CA1_Den_Current = data[4]
```

```python
with open('CA3_pyr_cell_LC.pkl', 'rb') as f: 
    data = pickle.load(f)
    CA3_TS  = data[0]
    CA3_Spikes = data[1]
    CA3_SpikeIndex = data[2]
    CA3_Som_Current = data[3]
    CA3_Den_Current = data[4]
```

```python
#Pyramidal Cell Raster
plt.figure(figsize=(20,10))
plt.plot(CA3_Spikes, CA3_SpikeIndex, '.k', label='CA3 Pyr. Cells')
plt.plot(CA1_Spikes, CA1_SpikeIndex, '.r', label='CA1 Pyr. Cells')
plt.xlabel('Time (ms)')
plt.ylabel('Neuron index')
plt.legend(loc="upper right")
```

```python
#Synaptic potentials
fig, axs = plt.subplots(2,figsize=(20,3))
axs[0].plot(CA1_TS,np.average(CA1_Som_Current,axis=0))
axs[1].plot(CA1_TS,np.average(CA1_Den_Current,axis=0))
```
